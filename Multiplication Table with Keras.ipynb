```python
import tensorflow as tf
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
from tensorflow.python.keras.callbacks import TensorBoard
from time import time
from datetime import datetime
```

```python
data_train = pd.read_csv('./dataset/Multiplication_Table_train.csv')
data_test = pd.read_csv('./dataset/Multiplication_Table_test.csv')
data_train[['a', 'b']].shape
data_test.head()
```

```python
def regressionModel():
    model=Sequential()
    model.add(Dense(128, input_dim=2, activation='sigmoid'))
    model.add(Dense(128, activation='sigmoid'))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam',  metrics=['acc'])
    return model
model=regressionModel()
model.summary()

```

# Tensorboard

```python
logdir="logs/scalars/" + datetime.now().strftime("%Y%m%d-%H%M%S")

tensorboard=TensorBoard(log_dir=logdir)
plot_model(model, to_file='./pictures/model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
new_data_test=zip(data_train.a, data_train.b)
split_train=pd.DataFrame(new_data_test, columns=['a','b'])
```

```python
history = model.fit(split_train, data_train.result, epochs=1000, callbacks=[tensorboard])
```
