```python
import tensorflow as tf
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
data_train = pd.read_csv('./dataset/Multiplication_Table_train.csv')
data_test = pd.read_csv('./dataset/Multiplication_Table_test.csv')
data_train[['a', 'b']].shape
```

```python
def regressionModel():
    model=Sequential()
    model.add(Dense(36, input_dim=2, activation='sigmoid'))
    model.add(Dense(36, activation='sigmoid'))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    return model
```

```python
model=regressionModel()
model.summary()
```

```python
new_data_test=zip(data_train.a, data_train.b)
split_train=pd.DataFrame(new_data_test, columns=['a','b'])
```

```python
model.fit(split_train, data_train.result, epochs=500)
```

```python
predictions=model.predict(split_train)
```

```python
new_data_test=list(zip(data_train.a, data_train.b, data_train.result, predictions))
df=pd.DataFrame(new_data_test, columns=['a','b', 'result' ,'predictions'])
df
```
